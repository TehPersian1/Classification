```python
import pandas as pd
import numpy as np
import sklearn as sk
import matplotlib.pyplot as plt

data1, data2, data3 = [], [], []
data1 = pd.read_csv('dataset1.csv')
data2 = pd.read_csv('dataset2.csv')
data3 = pd.read_csv('dataset3.csv')
data1.head()
```

# #1
in the cell above, we import the 3 datasets into lists and examine the contents of one to ensure everything has been imported correctly 

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier


data1Ys = data1[data1.columns[2]]
data1Xs = data1[data1.columns[0:2]]

gnb = GaussianNB()
gnb.fit(data1Xs, data1Ys)
data1gnbscore = gnb.score(data1Xs, data1Ys)

LR = LogisticRegression()
LR.fit(data1Xs, data1Ys)
data1LRscore = LR.score(data1Xs, data1Ys)

KNN = KNeighborsClassifier()
KNN.fit(data1Xs, data1Ys)
data1KNNscore = KNN.score(data1Xs, data1Ys)

print("GNB Score: ", data1gnbscore, "\nLR Score: ", data1LRscore, "\nKNN Score: ", data1KNNscore )
```

In the results of the code cells above, it is clear to see not only from the individual scores, but from the decision boundaries drawn up from each model that the radial basis function provides the best results when classifying our data, producing perfect results (score 1.0) on all 3 datasets.
